## 하나투어

```python
import requests              
from bs4 import BeautifulSoup
from urllib import request    
import os                     
        
# 데이터 추출    
url = "http://info.hanatour.com/?hanacode=main_TNB_getabout"
html = requests.get(url)
soup = BeautifulSoup(html.text, "lxml")

rankList = ".tabConts #hntxml_grp2 .rankingList"
RankImg = soup.select(rankList+' img')
RankTitle = soup.select(rankList+' .title')
RankDesk = soup.select(rankList+' .desc')

userName='YC'
folderURL ='C:/Users/'+userName+'/Desktop/hana/'

# 폴더 생성
if not os.path.exists(folderURL):
    os.makedirs(folderURL)    

hana_list = []
# 자료 저장
for i in range(len(RankImg)):        

    imgLink = RankImg[i].get("src")
    replace_ranktitle = RankTitle[i].text.replace("/"," ")
    savename =  folderURL + replace_ranktitle + ".jpg"
    
    request.urlretrieve(imgLink, savename)
    print("이미지 파일이 저장되었습니다...!")
    
    contents = {
        'Title':RankTitle[i].text,
        'Content':RankDesk[i].text,
    }
    
    hana_list.append(contents)
```

```python
import pandas as pd

# Dictionary를 DataFrame로 변환 후 csv파일로 저장
pd.DataFrame(hana_list).to_csv(folderURL+'hana_top4.csv', encoding='utf-8-sig')
    
# 저장한 csv파일 확인
hana = pd.read_csv(folderURL+'hana_top4.csv', index_col=0)
print(hana.shape)
hana.head(10)
```

## 롯데호텔

```python
from selenium import webdriver 
from bs4 import BeautifulSoup  

# 크롬 브라우저 열기
driver = webdriver.Chrome('C:/chromedriver.exe')

# 롯데호텔 룸 프로모션 화면으로 이동
url = "https://www.lottehotel.com/global/ko/hotel-offers/room-promotions.html"
driver.get(url)

# BeautifulSoup로 소스 받기
html = driver.page_source
soup = BeautifulSoup(html, "lxml")

# 데이터 추출
title = soup.select('.promotion__title')
date = soup.select('.promotion__date')
price = soup.select('.promotion-price__number')

lotte_list = []

for i in range(len(title)):
    
    contents = {
        'Title':title[i].text,
        'Date':date[i].text.strip().lstrip('예약 ,투숙 '),
        'Price':price[i].text,
    }
    
    lotte_list.append(contents)
```

```python
import pandas as pd

# Dictionary를 DataFrame로 변환 후 csv파일로 저장
pd.DataFrame(lotte_list).to_csv('C:/Users/YC/Desktop/lotte.csv', encoding='utf-8-sig')
    
# 저장한 csv파일 확인
lotte = pd.read_csv('C:/Users/YC/Desktop/lotte.csv', index_col=0)
print(lotte.shape)
lotte.head(10)
```

## 트립어드바이저

```python
from selenium import webdriver
from bs4 import BeautifulSoup

# 크롬 브라우저 열기
driver = webdriver.Chrome('C:/chromedriver.exe')

# 트립어드바이저 리뷰 화면으로 이동

    

url = f"https://www.tripadvisor.co.kr/Hotel_Review-g294197-d306130-Reviews-Lotte_Hotel_Seoul-Seoul.html"


driver.get(url)

ReviewID,ReviewTitle,ReviewText,ReviewDate = [],[],[],[]

# 1 ~ n번쨰 페이지까지 읽어옴
page = 5            
for i in range(page) :

    
    # BeautifulSoup로 소스 받아오기
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")

    # 리뷰글 : 작성자, 글제목, 내용, 숙박일
    ReviewID   .extend(soup.select('.social-member-event-MemberEventOnObjectBlock__event_type--3njyv span a'))
    ReviewTitle.extend(soup.select('.location-review-review-list-parts-ReviewTitle__reviewTitleText--2tFRT'))
    ReviewText .extend(soup.select('.location-review-review-list-parts-ExpandableReview__reviewText--gOmRC'))
    ReviewDate .extend(soup.select('.location-review-review-list-parts-EventDate__event_date--1epHa'))

    # 다음 페이지로 이동 버튼 클릭
    selector = ".ui_button.nav.next.primary "
    driver.find_element_by_css_selector(selector).click()
    

Review_list = []

for i in range(len(5,ReviewID)):

    Date = ReviewDate[i].text.lstrip('숙박 날짜 :')
    ID = ReviewID[i].text
    Title = ReviewTitle[i].text
    Text = ReviewText[i].text

    # Dictionary로 변환
    contents = {
        'Date':Date,
        'ID':ID,
        'Title':Title,
        'Text':Text
    }

    Review_list.append(contents)
```

```python
# pandas 임포트
import pandas as pd

# Dictionary를 DataFrame로 변환 후 csv파일로 저장
pd.DataFrame(Review_list).to_csv('C:/Users/YC/Desktop/trip.csv', encoding='utf-8-sig')
    
# 저장한 csv파일 확인
trip = pd.read_csv('C:/Users/YC/Desktop/trip.csv', index_col=0)
print(trip.shape)
trip.head(10)
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup 
import time

# 크롬 브라우저 열기
driver = webdriver.Chrome('C:/chromedriver.exe')

# 트립어드바이저 리뷰 화면으로 이동
url = "https://www.tripadvisor.co.kr/Hotel_Review-g294197-d306130-Reviews-Lotte_Hotel_Seoul-Seoul.html"
driver.get(url)

# 변수 초기화
page = 5
temp = ""
nextBtn = ".ui_button.nav.next.primary "
ReviewID,ReviewTitle,ReviewText,ReviewDate = [],[],[],[]

# 크롤링 반복
current_page = 0
while current_page < page :
    
    # 페이지 로드 후 게시글 작성자 확인
    soup = BeautifulSoup(driver.page_source, "lxml")
    checkID = soup.select('.social-member-event-MemberEventOnObjectBlock__event_type--3njyv span a')[0].text
    
    # 게시글 달라지면 저장
    if temp != checkID :                      
        
        # 리뷰글 : 작성자, 글제목, 내용, 숙박일
        ReviewID   .extend(soup.select('.social-member-event-MemberEventOnObjectBlock__event_type--3njyv span a'))
        ReviewTitle.extend(soup.select('.location-review-review-list-parts-ReviewTitle__reviewTitleText--2tFRT'))
        ReviewText .extend(soup.select('.location-review-review-list-parts-ExpandableReview__reviewText--gOmRC'))
        ReviewDate .extend(soup.select('.location-review-review-list-parts-EventDate__event_date--1epHa'))
    
        temp = checkID      # 현제 페이지 정보 갱신
        current_page += 1   # 현제 페이지 번호 갱신
    
    else :
        print("클릭 실패")
        time.sleep(1.5) # 게시글 이동 실패하면 1.5초 후 재시도
        
    # 다음 페이지 버튼 클릭
    driver.find_element_by_css_selector(nextBtn).click()
    
# 저장된 정보 출력
for i in range(len(ReviewID)):
    print(" #",i+1,':', ReviewDate[i].text.lstrip('숙박 날짜 :')) 
    print('이름 :', ReviewID[i].text) 
    print('제목 :', ReviewTitle[i].text)
    print('내용 :', ReviewText[i].text,'\n')
```
